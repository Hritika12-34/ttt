```python
#load dataset

import seaborn as sns
data= sns.load_dataset("iris")
data.head()
```

```python
#data cleaning

data.info()
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.nunique()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
#visualization

#1.distribution
import matplotlib.pyplot as plt

sns.histplot(data['petal_width'], kde=True)
plt.title("Petal_width Distribution")
plt.show()

sns.histplot(data['sepal_width'], kde=True)
plt.title("Sepal_width Distribution")
plt.show()
```

```python
#2.correlation

numeric_data= data.drop(columns=['species'])
correlation= numeric_data.corr()
sns.heatmap(correlation, annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#3.outliers

sns.boxplot(data=data.iloc[:,:-1])
plt.title("Outliers Check")
plt.xticks(rotation=45)
plt.show()
```

```python
#4.pairplot

sns.pairplot(data, hue='species')
plt.suptitle("Pairplot of Iris")
plt.show()
```
